### Problem Statement:

Compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference

```python
import numpy as np
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Visualize greyscale image
img = misc.face(gray=True)
plt.imshow(img, cmap=plt.cm.gray)
plt.show()
```

```python
plt.imsave("grey_image.png", img)
```

```python
img.shape
```

```python
row = img.shape[0]
col = img.shape[1]

c_img = img.reshape(row * col, 1)
kmeans = KMeans(n_clusters=5)
kmeans.fit(c_img)
```

```python
labels = kmeans.labels_
labels = labels.reshape(row, col)
plt.imshow(labels)
plt.show()
```

```python
plt.imsave("compressed_image.png", labels)
```

```python
plt.imshow(img, cmap=plt.cm.gray)
plt.show()

plt.imshow(labels)
plt.show()
```
